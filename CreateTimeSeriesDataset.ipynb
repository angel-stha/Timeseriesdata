```python
import pandas as pd
import numpy as np
import random
import datetime
```

```python
data = pd.DataFrame()
```

```python
'''Creating data for sale of TVremote house for a year on daily basis.
    Adding timestamp to data; as TV remote house can have multiple
    item and there can be sale of any times in each day. So randomly
    creating multiple timestamp per day'''
date = datetime.datetime(2020,1,1)
b=0
for i in range(398): 
    date += datetime.timedelta(days=1)
    a=(random.randint(1,9))
    for j in range(a):
        data.loc[b+j,'timestamp']=(date)
    b=b+j
```

```python
data
# Filling up the items choosing randomly from 9 items in shop
```

```python
# Adding items randomly for TVhouse
items = ["DTH remote","Dth reciever","LMB","LED remote","TV bracket","Videojack 3 pin","Video jack 2 pin","solder","iron"]
for i in range(1599):
    data.loc[i,'item_id']=(random.choice(items))
    
```

```python
# Now adding price tag, quantity and sales per day
data['price']=pd.DataFrame(np.random.randint(50,250,size=(1599, 1)))
data["quantity_sold"] = pd.DataFrame(np.random.randint(100,5000,size=(1599, 1)))
data["sales_per_day"] = (data["price"]) * (data["quantity_sold"])
```
